```python
import pandas as pd
titanic_df = pd.read_csv('/content/drive/MyDrive/Colab/train.csv')
male_df = titanic_df[titanic_df['Sex']=='male']
male_names=male_df['Name'].str.extract(r'(?:Mr. |Don. |Master. )\s*(\S+)') #имя - первое слово после мистер/мастер и до пробела
female_df = titanic_df[titanic_df['Sex'] == 'female']
female_names=female_df['Name'].str.split().str[-1].str.extract(r'([A-Za-z]+)') #имя - посдеднее слово слово в столбце
most_popular_female_name=female_names.mode()
most_popular_male_name=male_names.mode()
print(male_names[0].value_counts()) #проверка количества повторов мужских имен
print(female_names[0].value_counts()) #проверка количества повторов женских имен
print("Самое частое мужское имя - " + most_popular_male_name)
print("Самое частое женское имя - " + most_popular_female_name)
```

```python
import pandas as pd
titanic_df = pd.read_csv('/content/drive/MyDrive/Colab/train.csv')
female_df = titanic_df[titanic_df['Sex'] == 'female']
female_df1 = female_df[female_df['Pclass']==1]
female_names1=female_df1['Name'].str.split().str[-1].str.extract(r'([A-Za-z]+)') #имя - посдеднее слово слово в столбце
fname1=female_names1.mode()
female_df2 = female_df[female_df['Pclass']==2]
female_names2=female_df2['Name'].str.split().str[-1].str.extract(r'([A-Za-z]+)') #имя - посдеднее слово слово в столбце
fname2=female_names2.mode()
female_df3 = female_df[female_df['Pclass']==3]
female_names3=female_df3['Name'].str.split().str[-1].str.extract(r'([A-Za-z]+)') #имя - посдеднее слово слово в столбце
fname3=female_names3.mode()
print("Cамое популярное женское имя среди пассажирок 1 класса: "+fname1.to_string(index=False))
print("Cамое популярное женское имя среди пассажирок 2 класса: "+fname2.to_string(index=False))
print("Cамое популярное женское имя среди пассажирок 3 класса: "+fname3.to_string(index=False))
```

```python
import pandas as pd
titanic_df = pd.read_csv('/content/drive/MyDrive/Colab/train.csv')
male_df = titanic_df[titanic_df['Sex'] == 'male']
male_df1 = male_df[male_df['Pclass']==1]
male_names1=male_df1['Name'].str.extract(r'(?:Mr. |Don. |Master. )\s*(\S+)') #имя - посдеднее слово слово в столбце
mname1=male_names1[0].mode()
male_df2 = male_df[male_df['Pclass']==2]
male_names2=male_df2['Name'].str.extract(r'(?:Mr. |Don. |Master. )\s*(\S+)') #имя - посдеднее слово слово в столбце
mname2=male_names2.mode()
male_df3 = male_df[male_df['Pclass']==3]
male_names3=male_df3['Name'].str.extract(r'(?:Mr. |Don. |Master. )\s*(\S+)') #имя - посдеднее слово слово в столбце
mname3=male_names3.mode()
print("Cамое популярное мужское имя среди пассажиров 1 класса: "+mname1.to_string(index=False))
print("Cамое популярное мужское имя среди пассажиров 2 класса: "+mname2.to_string(index=False))
print("Cамое популярное мужское имя среди пассажиров 3 класса: "+mname3.to_string(index=False))
```

```python
import pandas as pd
titanic_df = pd.read_csv('/content/drive/MyDrive/Colab/train.csv')
cabincount = titanic_df['Cabin'].value_counts()
#поочередная печать кабин с 2, 3 и 4 пассажирами:
cabinsfor2=cabincount[cabincount==2]
cabinsfor3=cabincount[cabincount==3]
cabinsfor4=cabincount[cabincount==4]
print('Каюты с 2 пассажирами:', ', '.join(cabinsfor2.index.astype(str)))
print('Каюты с 3 пассажирами:',', '.join(cabinsfor3.index.astype(str)))
print('Каюты с 4 пассажирами:',', '.join(cabinsfor4.index.astype(str)))
```
