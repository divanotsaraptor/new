```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import math
df=pd.read_csv('/content/drive/MyDrive/Colab/train.csv')

# survived
surv=df['Survived'].value_counts()
plt.ylabel(' , ')
surv.plot.bar(x='Survived')
plt.title('  ')
plt.xticks(ticks=[0,1], labels=[' ', ''], rotation=0)
plt.show()

#  
pclasses=df['Pclass'].value_counts()
pclasses.plot.bar(x='Pclass')
plt.title('  ')
plt.xticks(rotation=0)
plt.ylabel(' , ')
plt.show()

#  
ages=df['Age'].dropna().apply(lambda x: math.ceil(x)).value_counts()
ages.plot.hist(x='Age')
plt.title('   ')
plt.xscale('log',)
plt.xticks(rotation=90)
plt.xlabel(' , ')
plt.ylabel(' , ')
plt.xlim(1,100)
plt.ylim(0,20)
plt.show()

#   
bysex=df['Sex'].value_counts()
bysex.plot.bar(x='Sex')
plt.title('  ')
plt.ylabel(' , ')
plt.xticks(rotation=0)
plt.show()

#   parch
byparch=df['Parch'].dropna().value_counts()
byparch.plot.bar(x='Parch')
plt.ylabel(' , ')
plt.show()

#boxplot    
df.boxplot(column='Age')
plt.title('   ')
plt.ylabel(', ')
plt.show()

#    
byclass= df['Pclass'].value_counts()
byclass.plot(kind='pie', autopct='%1.1f%%')
plt.title('    ', fontsize=9)
plt.ylabel('')
plt.show()

#  
bysurvived= df['Survived'].value_counts()
bysurvived.plot(kind='pie', autopct='%1.1f%%')
plt.title(' ')
plt.ylabel('')
plt.show()



#  
fig = px.sunburst(df, path=['Sex', 'Pclass'],
                  branchvalues='total',
                  title='  ')

fig.show()
```

```python
#pairplot   
sns.pairplot(df, plot_kws={'alpha':0.2})
plt.suptitle('    ', fontsize=32)
plt.tight_layout()
plt.show()
```
